```python
import pandas as pd
import requests 
import json
from sqlalchemy import create_engine
import psycopg2 

import urllib.parse
```

```python
csv_file="Restaurants/USObesebyStates.csv"
obese_data_df= pd.read_csv(csv_file)
obese_data_df.head()
```

```python
obese_data_df.rename(columns={"United States" : "State"}, inplace = True)
obese_data_df
```

```python
from bs4 import BeautifulSoup
import requests
website = requests.get('https://www.nicerx.com/fast-food-capitals/')
soup = BeautifulSoup(website.content, 'html.parser')
id = soup.find("table", {"class" : "tableizer-table"}).find_all("tr")
print(id)
```

```python
stateOb_df = []
for row in id:
#      print(row)
    # Find all data for each column
    columns = row.find_all('td')
#     print(len(columns))
    if(columns != []):
#         print(columns[2].text)
        State = columns[0].text
        All_fast_food_restaurants = columns[1].text
        Full_service_restauants = columns[2].text
        Subway = columns[3].text
        Starbucks = columns[4].text
        McDonalds = columns[5].text
        Dunkin_Donut = columns[6].text
        Burger_King = columns[7].text
        Taco_Bell = columns[8].text
        Dominos = columns[9].text
        Wendys = columns[10].text
        Dairy_Queen = columns[11].text
        KFC = columns[12].text
        statedict = {'State': State,'All fast food restaurants': All_fast_food_restaurants,'Full_service restauants': Full_service_restauants,
                                         'Subway': Subway,
                        'Starbucks': Starbucks, 'McDonalds': McDonalds,'Dunkin Donut': Dunkin_Donut, 'Burger King' : Burger_King,
                                        'Taco Bell': Taco_Bell, 'Dominos' : Dominos, 'Wendys' : Wendys, 'Dairy Queen' : Dairy_Queen,
                                        'KFC' : KFC}
#         print(statedict)
        stateOb_df.append(statedict)
#         'All fast food restaurants': All fast food restaurants,'Full-service restauants': Full-service restauants,'Dunkin Donut': Dunkin Donut,
# print(stateOb_df)
results_df = pd.DataFrame(stateOb_df)
results_df
```
